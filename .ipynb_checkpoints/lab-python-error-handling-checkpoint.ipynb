## Exercise: Error Handling for Managing Customer Orders

The implementation of your code for managing customer orders assumes that the user will always enter a valid input. 

For example, we could modify the `initialize_inventory` function to include error handling.
   - If the user enters an invalid quantity (e.g., a negative value or a non-numeric value), display an error message and ask them to re-enter the quantity for that product.
   - Use a try-except block to handle the error and continue prompting the user until a valid quantity is entered.

```python
# Step 1: Define the function for initializing the inventory with error handling
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory

# Or, in another way:

def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_input = False
        while not valid_input:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity >= 0:
                    inventory[product] = quantity
                    valid_input = True
                else:
                    print("Quantity cannot be negative. Please enter a valid quantity.")
            except ValueError:
                print("Invalid input. Please enter a valid quantity.")
    return inventory
```

````python
# Necessary Definitions

products = ['t-shirt', 'mug', 'hat', 'book', 'keychain']

inventory = {}

customer_orders = set()

prices = []
````

````python
# Inventory function and calling of function

def initialize_inventory(products):
    inventory = {}
    for product in products:
        while True:
            try:
                value = int(input(f'How many items of {product} is in inventory? '))
                if value < 0:
                    print('Inventory cannot be negative. Please enter a valid number.')
                else:
                    inventory[product] = value
                    break  # valid input, move to next product
            except ValueError:
                print('Invalid input. Please enter a number.')
    return inventory

inventory = initialize_inventory(products)
````

````python
# Modify the get_customer_orders function to include error handling.

def get_customer_orders():
    while True:
        try:
            many_products = int(input('How many products does the client want to order? '))
            if many_products <= 0:
                print('Invalid input. Please enter a positive number.')
            else:
                break  # valid number, exit loop
        except ValueError:
            print('Invalid input. Please enter a number.')
  
    for quantity in range(many_products): 
        item_order = str(input('Which item client want to order?'))
        while item_order not in products: # if a product is not in stock (either by typos or 0 stock), then it is not in the inventory dict
            print('Invalid product. Write an item from products')
            item_order = str(input('Which item client want to order?'))
        customer_orders.add(item_order)
    print('Customer Orders:', customer_orders)
    print('Order Finished')  
    return customer_orders

customer_orders = get_customer_orders() 
````

````python
# Modify the calculate_total_price function to include error handling.

def total_price(customer_orders):
    for product in customer_orders:
        while True:
            try:
                price = float(input(f'Unit price for {product}: '))
                if price <= 0:                          
                    print('Price cannot be negative nor 0. Please input correct price')
                else:
                    prices.append(price)
                    break                              
            except ValueError:                         
                print('Invalid price. Please enter a number.')
    return sum(prices)

total_price(customer_orders)

print(f'Total order price: {sum(prices)}')
````
